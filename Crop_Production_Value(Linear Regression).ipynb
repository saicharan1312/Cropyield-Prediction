### The area of the land,the type of crop and season determines the production value.
### Hence,it is a Multpile Linear Regression.

```python
Season=pd.get_dummies(df["Season"])
Season
```

```python
Crops=pd.get_dummies(df["Crop"])
Crops
```

```python
d=pd.get_dummies(df)
```

```python
df=pd.concat([df,Season,Crops],axis=1)
#d.head(3)
df.head(3)
```

```python
df.drop(["Season","Crop"],axis=1,inplace=True)
df.head(3)
```

```python
#Split the data into X,y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=d.drop("Production",axis=1)
y=d[["Production"]]
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.50,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
reg=linear_model.LinearRegression()
```

```python
reg.fit(X_train,y_train)
```

```python
prod_value=reg.predict(X_test)
```

```python
prod_value
```

```python
from sklearn.metrics import r2_score
```

```python
r2_score(y_test,prod_value)
```
